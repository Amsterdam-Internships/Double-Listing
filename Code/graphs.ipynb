```python
#Create an average of the 5 runs of the active learning set up 

f_score_nomod_rf_ams = []
f_score_img_rf_ams = []
f_score_nlp_rf_ams = []


for dicts in no_modal_results_full[0][:4]:
    f_score_nomod_rf_ams.append(dicts['f_score'])
    
f_score_nomod_rf_mean_ams =np.mean(f_score_nomod_rf_ams,axis=0)
f_score_nomod_rf_sd_ams = np.std(f_score_nomod_rf_ams,axis=0)

for dicts in img_results_full:
    f_score_img_rf_ams.append(dicts['f_score'])
    
f_score_img_rf_mean_ams =np.mean(f_score_img_rf_ams,axis=0)
f_score_img_rf_sd_ams = np.std(f_score_img_rf_ams,axis=0)

for dicts in nlp_results_full:
    f_score_nlp_rf_ams.append(dicts['f_score'])
    
f_score_nlp_rf_mean_ams =np.mean(f_score_nlp_rf_ams,axis=0)
f_score_nlp_rf_sd_ams = np.std(f_score_nlp_rf_ams,axis=0)
```

```python
#Baselines
fig, ax = plt.subplots(1)
fig.set_size_inches(10,8)
ax.plot(np.arange(101), f_score_full_rf_mean, lw=1, label='boot rf', color='blue')
#ax.plot(np.arange(101), f_score_full_cmt_mean, lw=1, label='boot committee', color='red')
ax.plot(np.arange(101), f_score_no_boot_rf_mean, lw=1, label=' no boot rf', color='green')
ax.plot(np.arange(101), f_score_rf_noboot_nowarm_mean, lw=1, label='no boot no warm rf', color='purple')
ax.plot([0, 101], [0.7961165048543688,0.7961165048543688], 'k-', lw=2, label= 'SVM supervised')
ax.plot([0, 101], [0.7961165048543688,0.7961165048543688], 'k-', lw=2, label= 'Decision tree supervised')
ax.fill_between(np.arange(101), f_score_full_rf_mean + ((f_score_full_rf_sd/np.sqrt(5))*2.776), f_score_full_rf_mean-((f_score_full_rf_sd/np.sqrt(5))*2.776)
                ,facecolor='blue', alpha=0.2)
ax.fill_between(np.arange(101), f_score_no_boot_rf_mean + ((f_score_no_boot_rf_sd/np.sqrt(5))*2.776), f_score_no_boot_rf_mean-((f_score_no_boot_rf_sd/np.sqrt(5))*2.776)
                ,facecolor='green', alpha=0.1)
ax.fill_between(np.arange(101), f_score_rf_noboot_nowarm_mean + ((f_score_rf_noboot_nowarm_sd/np.sqrt(5))*2.776), f_score_rf_noboot_nowarm_mean-((f_score_rf_noboot_nowarm_sd/np.sqrt(5))*2.776)
                ,facecolor='purple', alpha=0.1)                                
ax.set_title('F score and 95% CI for Web Commons Project data')
ax.legend(loc='lower right')
ax.set_xlabel('Iterations')
ax.set_ylabel('F_score ')

ax.grid()
```
